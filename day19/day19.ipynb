```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Load dataset
data = pd.read_csv("fraud_data.csv")

# Convert Yes/No into numbers
data['Fraud'] = data['Fraud'].map({
    'Yes': 1,
    'No': 0
})

# Input and output
X = data[['Amount']]
y = data['Fraud']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# XGBoost
xgb = XGBClassifier()
xgb.fit(X_train, y_train)

# Compare accuracy
print("Random Forest Score:")
print(rf.score(X_test, y_test))

print("\nXGBoost Score:")
print(xgb.score(X_test, y_test))
```

# Performance Comparison

- Random Forest performed well with stable predictions.
- XGBoost improved accuracy by learning from previous mistakes.
- Boosting models can achieve higher performance than bagging methods.

## Tradeoffs
- XGBoost is more powerful but slightly more complex.
- Training time can be higher for boosting models.
