```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load dataset
data = pd.read_csv("creditcard.csv")

# Create a categorical column
data['TransactionType'] = ['Online' if x > 100 else 'Offline'
                           for x in data['Amount']]

# BEFORE Encoding
print("Dataset Before Encoding:")
print(data[['Amount', 'TransactionType']].head())

# Label Encoding
label = LabelEncoder()

data['TransactionType_Label'] = label.fit_transform(
    data['TransactionType']
)

# One-Hot Encoding
encoded_data = pd.get_dummies(
    data,
    columns=['TransactionType']
)

# AFTER Encoding
print("\nDataset After Encoding:")
print(encoded_data.head())

# Features and Target
X = encoded_data[['Amount', 'TransactionType_Label']]
y = encoded_data['Class']

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# Train Model
model = RandomForestClassifier()

model.fit(X_train, y_train)

# Accuracy
score = model.score(X_test, y_test)

print("\nModel Accuracy:")
print(score)

# Save Encoded Dataset
encoded_data.to_csv("encoded_creditcard.csv", index=False)

print("\nEncoded dataset saved!")
```

# Performance Comparison Report

- Before encoding, the dataset contained text values like Online and Offline.
- Label Encoding converted text into numbers.
- One-Hot Encoding created separate columns for categories.
- Encoded features improved Machine Learning compatibility.
- Random Forest model performed successfully after encoding.
