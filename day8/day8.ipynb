```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("sales_data_sample.csv", encoding='latin1')

# Show first 5 rows
print(data.head())

# Summary statistics
print(data.describe())

# Sales distribution
plt.hist(data['SALES'])
plt.title("Sales Distribution")
plt.show()

# Quantity vs Sales
plt.scatter(data['QUANTITYORDERED'], data['SALES'])
plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.show()
```

# Business Insights

1. Most sales are in the medium range.
2. Higher quantity orders generally lead to higher sales.
3. Some sales values are extremely high compared to others.
4. Customers from different countries contribute to sales.
5. Sales activity changes across months and years.
