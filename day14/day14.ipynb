```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("feature_engineered_data.csv")

# -----------------------------
# BEFORE removing feature
# -----------------------------
X1 = data[['QUANTITYORDERED']]
y = data['SALES']

X_train1, X_test1, y_train1, y_test1 = train_test_split(
    X1, y, test_size=0.2, random_state=42
)

model1 = LinearRegression()
model1.fit(X_train1, y_train1)

score_before = model1.score(X_test1, y_test1)

print("Performance Before Removing Feature:")
print(score_before)

# -----------------------------
# AFTER removing important feature
# -----------------------------
X2 = data[['PRICEEACH']]

X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X2, y, test_size=0.2, random_state=42
)

model2 = LinearRegression()
model2.fit(X_train2, y_train2)

score_after = model2.score(X_test2, y_test2)

print("\nPerformance After Removing Feature:")
print(score_after)
```
