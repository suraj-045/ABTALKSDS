```python
import pandas as pd

# Create model comparison table
models = {
    'Model': [
        'Logistic Regression',
        'KNN',
        'Decision Tree',
        'Random Forest',
        'XGBoost'
    ],

    'Accuracy': [
        0.80,
        0.75,
        0.85,
        1.00,
        0.83
    ]
}

# Create dataframe
df = pd.DataFrame(models)

# Show comparison
print(df)

# Best model
best_model = df.loc[df['Accuracy'].idxmax()]

print("\nBest Model:")
print(best_model)
```

# Best Model Analysis

## Selected Model
Random Forest

## Why It Performed Best
- Highest accuracy
- Better generalization
- Reduced overfitting
- Stable predictions

## Weakness
- More complex than Logistic Regression
- Harder to interpret than Decision Trees
