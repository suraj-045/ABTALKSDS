```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("cleaned_sales_data.csv")

# Show first 5 rows
print(data.head())

# Plot sales distribution
plt.hist(data['SALES'])
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Count")
plt.show()

# Plot quantity distribution
plt.hist(data['QUANTITYORDERED'])
plt.title("Quantity Ordered")
plt.xlabel("Quantity")
plt.ylabel("Count")
plt.show()

# Find highest sales
print("Highest Sales:")
print(data['SALES'].max())

# Find lowest sales
print("Lowest Sales:")
print(data['SALES'].min())
```

## Insights

1. Most sales values are between medium ranges.
2. Some orders have very high sales.
3. Quantity ordered varies across customers.
4. Dataset contains multiple countries.
5. Sales data changes across months and years.
