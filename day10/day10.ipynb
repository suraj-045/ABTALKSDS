```python
import pandas as pd

# Load dataset
data = pd.read_csv("cleaned_sales_data.csv")

# 1. Identify features
# --------------------------
print("Numerical Columns:")
print(data.select_dtypes(include=['int64', 'float64']).columns)

print("\nCategorical Columns:")
print(data.select_dtypes(include=['object']).columns)

# 2. Encoding (convert text to numbers)
# --------------------------
data = pd.get_dummies(data, columns=['STATUS', 'DEALSIZE'])

# 3. Scaling (simple normalization)
# --------------------------
data['SALES'] = data['SALES'] / data['SALES'].max()

# 4. Create new features
# --------------------------
data['PRICE_PER_UNIT'] = data['SALES'] / data['QUANTITYORDERED']
data['HIGH_QUANTITY'] = data['QUANTITYORDERED'] > 40

# Show result
# --------------------------
print("\nUpdated Data:")
print(data.head())

# Save dataset
# --------------------------
data.to_csv("feature_engineered_data.csv", index=False)

print("\nFeature engineered dataset saved!")
```
