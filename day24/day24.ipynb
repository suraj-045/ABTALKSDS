```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load dataset
data = pd.read_csv("creditcard.csv")

# Features and target
X = data.drop('Class', axis=1)
y = data['Class']

# -----------------------------
# BEFORE PCA
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

model = RandomForestClassifier()

model.fit(X_train, y_train)

before_score = model.score(X_test, y_test)

print("Accuracy Before PCA:")
print(before_score)

# Apply PCA
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

# AFTER PCA
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X_pca, y, test_size=0.2
)

model2 = RandomForestClassifier()

model2.fit(X_train2, y_train2)

after_score = model2.score(X_test2, y_test2)

print("\nAccuracy After PCA:")
print(after_score)

# Explained Variance
print("\nExplained Variance Ratio:")

# Visualization
plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y,
    s=5
)

plt.title("PCA Visualization")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```

# PCA Variance Analysis

- PCA reduced dataset dimensions to 2 components.
- Most important information was preserved after compression.
- Model accuracy remained strong after dimensionality reduction.
- PCA simplified the dataset and reduced complexity.
