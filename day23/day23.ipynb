```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load dataset
data = pd.read_csv("creditcard.csv")

# Features and target
X = data[['V14', 'V17', 'Amount']]
y = data['Class']

# -----------------------------
# BEFORE Feature Selection
# -----------------------------
X_before = data[['V14', 'V17', 'Amount', 'V10', 'V12']]

X_train, X_test, y_train, y_test = train_test_split(
    X_before, y, test_size=0.2
)

model = RandomForestClassifier()

model.fit(X_train, y_train)

before_score = model.score(X_test, y_test)

print("Accuracy Before Feature Selection:")
print(before_score)

# -----------------------------
# AFTER Feature Selection
# -----------------------------
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X, y, test_size=0.2
)

model2 = RandomForestClassifier()

model2.fit(X_train2, y_train2)

after_score = model2.score(X_test2, y_test2)

print("\nAccuracy After Feature Selection:")
print(after_score)

# -----------------------------
# Important Features
# -----------------------------
print("\nSelected Important Features:")
print(X.columns)
```

# Feature Selection Report

- Model achieved high accuracy before and after feature selection.
- Selected features: V14, V17, and Amount.
- Feature selection reduced complexity with minimal accuracy loss.
- Important features improved model efficiency and interpretability.
