```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load dataset
data = pd.read_csv("movies.csv")

# Remove text columns
X = data.drop(['Movie', 'Genre'], axis=1)

# Create model
model = NearestNeighbors(n_neighbors=3)

# Train model
model.fit(X)

# Find similar movies
distances, indices = model.kneighbors(X.iloc[[0]])

print("Recommended Movies:")

for i in indices[0]:
    print(data.iloc[i]['Movie'])
```

# K Value Comparison

- K=2 gave very specific recommendations
- K=5 gave broader recommendations
- Best K = 3 because balance between accuracy and variety
